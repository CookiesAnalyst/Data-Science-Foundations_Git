```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/sergi/Desktop/MBDA/Master/Data Science Foundations/Session_2/2_titanic_dataset.csv")
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[12,10])
sns.countplot(df['Survived'])
plt.show()
```

```python
plt.figure(figsize=[12,10])
sns.barplot('Sex','Survived', data=df)

#show the plot
plt.show ()
```

```python
sns.pointplot(x='Parch', y='Survived', hue='Sex', data=df, palette={'male':'blue', 'female':'red'})
plt.show()
```

```python
plt.figure(figsize=[12,10])
sns.swarmplot (x='Sex', y='Age', data=df)
plt.show()
```

```python
plt.figure(figsize=[12,10])
sns.factorplot ('Pclass', 'Survived', 'Sex', data=df, kind='bar', palette='muted', legend=False)
plt.show()
```

```python
plt.figure(figsize=[12,10])
sns.factorplot ('Pclass', 'Survived', 'Sex', data=df, kind='bar', palette='muted', legend=True)
plt.show()
```

```python
g = sns.FacetGrid (df, col='Sex', row = 'Pclass', margin_titles = True)
g.map(plt.scatter, 'Fare', 'Age')

#Add a tittle to the figure
g.fig.suptitle ("scatter plot")

#show the plot
plt.show()
```
